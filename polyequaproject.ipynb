```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
e = 0.1
t = 0.0
x = 1.0
v = 0.0
a = -1.0
m = 60
time = []
position = []
velocity = []
acceleration = []
time.append(t)
position.append(x)
velocity.append(v)
acceleration.append(a)
print("t =", t,"x = ", x,"v = ", v,"a = ", a)
v = v + (e/2)*a
print("v = ", v)
for i in range(1,m):
    x = x + e*v
    a = -x
    v = v - e*x
    t = t + e
    time.append(t)
    position.append(x)
    velocity.append(v)
    acceleration.append(a)
    print("t = ", t, "x = ", x, "a = ", a)
    print("v = ", v)
plt.plot(time, position, 'o')
plt.plot(time, velocity, 'o')
plt.plot(time, acceleration, 'o')
plt.title("Simple Harmonic Motion with k/m = 1")
plt.xlabel("Time")
plt.ylabel("Position, Velocity, Acceleration")
plt.legend(["Position", "Velocity", "Acceleration"], loc = "lower right")
plt.show()
```
